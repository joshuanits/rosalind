```python
punnets = {
    ('AA', 'AA'): 1.0,
    ('AA', 'Aa'): 1.0,
    ('AA', 'aa'): 1.0,
    ('Aa', 'Aa'): 0.75,
    ('aa', 'aa'): 0.0,
    ('Aa', 'aa'): 0.5,
}

def inheritance(genotype_a, genotype_b, is_dominant=True):
    key = tuple(sorted((genotype_a, genotype_b)))

    prob_dominant = punnets.get(key)

    if prob_dominant is None:
        raise ValueError(f"Invalid genotype combination: {key}")
    
    return prob_dominant if is_dominant else 1 - prob_dominant
```

```python
input_text = "18931 18976 18794 16320 18055 19661"
populations = input_text.split(" ")

genotypes = [
    "AA-AA",
    "AA-Aa",
    "AA-aa",
    "Aa-Aa",
    "Aa-aa",
    "aa-aa"
]

result = 0

for i in range(6):
    genotype = tuple(genotypes[i].split("-"))
    n = int(populations[i])

    print(n)

    result += n * inheritance(genotype[0], genotype[1]) * 2

print(result)
```
