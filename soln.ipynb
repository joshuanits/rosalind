```python
def read_input(file):
    with open(f"rosalind_{file}.txt") as f:
        return f.read()[:-1]
```

```python
# rna

input = read_input("rna")

output = ""

for char in input:
    mapped = "U" if char == "T" else char
    output += mapped

print(output)
```

```python
# revc

input = read_input("revc")

output = ""

complements = {
    "A": "T",
    "T": "A",
    "G": "C",
    "C": "G",
}

for char in input:
    output = complements[char] + output

print(output)
```

```python
# fib

input = read_input("fib")

months, per_litter = [int(x) for x in input.split(" ")]

population = []

for i in range(months):
    if i < 2:
        pop = 1
    else:
        pop = population[-1] + population[-2] * per_litter

    population.append(pop)

print(population)
print(population[-1])
```

```python
def parse_fasta(text):
    result = []

    current_id = None
    current_sequence = None
    reading_id = False

    for char in text:
        if char == ">":
            if current_id:
                result.append((current_id, current_sequence))

            current_id = ""
            current_sequence = ""
            reading_id = True

            continue

        if char == "\n":
            reading_id = False
            continue

        if reading_id:
            current_id += char
            continue

        current_sequence += char

    result.append((current_id, current_sequence))

    return result

def count_symbols(sequence):
    counts = {'A': 0, 'C': 0, 'G': 0, 'T': 0}
    for nucleotide in sequence:
        if nucleotide in counts:
            counts[nucleotide] += 1

    return counts
```

```python
# gc 

from rosalind import FASTA, Sequence
input = read_input("gc")
data = FASTA.parse(input)

max_id = None
max_value = -1

for id, sequence in data.items():
 
    value = sequence.gc_ratio

    if value > max_value:
        max_id = id
        max_value = value

print(max_id)
print(round(max_value, 7))
```
